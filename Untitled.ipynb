```python
import json
import urllib.request

# Function to print keys recursively
def print_keys(data, parent_key=""):
    if isinstance(data, dict):
        for key, value in data.items():
            new_key = f"{parent_key}.{key}" if parent_key else key
            print(new_key)
            print_keys(value, parent_key=new_key)
    elif isinstance(data, list):
        for index, item in enumerate(data):
            new_key = f"{parent_key}[{index}]"
            print(new_key)
            print_keys(item, parent_key=new_key)

# URL to the JSON data
url = "https://2u-data-curriculum-team.s3.amazonaws.com/dataviz-classroom/v1.1/14-Interactive-Web-Visualizations/02-Homework/samples.json"

# Fetch the JSON data
with urllib.request.urlopen(url) as response:
    data = json.load(response)

# Print all keys recursively
print_keys(data)
```

```python
import json
import urllib.request

# Function to print keys recursively
def print_keys(data, parent_key="", key_counts=None):
    if key_counts is None:
        key_counts = {}

    if isinstance(data, dict):
        for key, value in data.items():
            new_key = f"{parent_key}.{key}" if parent_key else key
            key_counts[new_key] = key_counts.get(new_key, 0) + 1
            print(new_key, "-", key_counts[new_key])
            print_keys(value, parent_key=new_key, key_counts=key_counts)
    elif isinstance(data, list):
        if all(isinstance(item, dict) for item in data):
            print_keys(data[0], parent_key=parent_key, key_counts=key_counts)
        else:
            key_counts[parent_key] = key_counts.get(parent_key, 0) + 1
            print(parent_key, "-", key_counts[parent_key])

# URL to the JSON data
url = "https://2u-data-curriculum-team.s3.amazonaws.com/dataviz-classroom/v1.1/14-Interactive-Web-Visualizations/02-Homework/samples.json"

# Fetch the JSON data
with urllib.request.urlopen(url) as response:
    data = json.load(response)

# Print all keys recursively with count
print_keys(data)
```

```python
import json
import urllib.request

# Function to print keys recursively
def print_keys(data, parent_key="", key_counts=None, printed_keys=None):
    if key_counts is None:
        key_counts = {}
    if printed_keys is None:
        printed_keys = set()

    if isinstance(data, dict):
        for key, value in data.items():
            new_key = f"{parent_key}.{key}" if parent_key else key
            key_counts[new_key] = key_counts.get(new_key, 0) + 1
            if new_key not in printed_keys:
                print(new_key, "-", key_counts[new_key])
                printed_keys.add(new_key)
            print_keys(value, parent_key=new_key, key_counts=key_counts, printed_keys=printed_keys)
    elif isinstance(data, list):
        for item in data:
            print_keys(item, parent_key=parent_key, key_counts=key_counts, printed_keys=printed_keys)

# URL to the JSON data
url = "https://2u-data-curriculum-team.s3.amazonaws.com/dataviz-classroom/v1.1/14-Interactive-Web-Visualizations/02-Homework/samples.json"

# Fetch the JSON data
with urllib.request.urlopen(url) as response:
    data = json.load(response)

# Print all keys recursively with count
print_keys(data)

```
